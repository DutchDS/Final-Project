```python
#Import Dependencies
import pandas as pd
```

```python
covid_states_csv = "../assets/data/us_states_covid19_daily.csv"
read_covid = pd.read_csv(covid_states_csv)

read_covid.dropna()
read_covid.head()
```

```python
#Drop some columns
new_covid_df = read_covid[['date','state','positive']]
new_covid_df = pd.to_datetime(read_covid['date'],format='%Y%m%d')
```

```python
date_list = []
for i in new_covid_df:
    date_list.append(i)
    print(i)
```

```python
date_dict = {'date':date_list, 'state': read_covid['state'],'total': read_covid['total']}

date_df = pd.DataFrame(date_dict)
```

```python
date_df
```

```python
covid_states_pivot = date_df.pivot(index='state',values='total',columns='date')
```
