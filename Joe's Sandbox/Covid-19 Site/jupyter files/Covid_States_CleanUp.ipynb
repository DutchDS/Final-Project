```python
#Import Dependencies
import pprint as pp
import pandas as pd
```

```python
covid_states_csv = "../assets/data/us_states_covid19_daily.csv"
read_covid = pd.read_csv(covid_states_csv)

read_covid.dropna()
read_covid.head()
```

```python
#Drop some columns
new_covid_df = read_covid[['date','state','positive']]
new_covid_df = pd.to_datetime(read_covid['date'],format='%Y%m%d')
```

```python
date_list = []
for i in new_covid_df:
    date_list.append(i)
    print(i)
```

```python
date_dict = {'state': read_covid['state'],'date': read_covid['date'] ,'formatted_date':date_list,'total': read_covid['total']}

date_df = pd.DataFrame(date_dict)
date_df = date_df.fillna(0)
int_date = date_df.date.astype(int)
int_values = date_df.total.astype(int)

new_date_dict = {'state': read_covid['state'],'date': int_date ,'formatted_date': date_list,'total': int_values}
new_date_df = pd.DataFrame(new_date_dict)
new_date_df

df_group = new_date_df.groupby(['state','date'])
state_total = df_group['total'].sum()
state_dict = {'total':state_total}
df = pd.DataFrame(state_dict).reset_index()
df = df.dropna(0)
df.max()
```

```python
df.to_csv("states_total.csv", index=False)
new_date_df.to_csv("non_pivot_states.csv", index=False)
```
